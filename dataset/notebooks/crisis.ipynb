```python
replace_map ={
    "bomber":"bomb",
    "bombing":"bomb",

    "bushfire":"fire",
    "firefighters":"fire",

    "dead":"death",
    "deadly":"death",

    "floodwaters":"flood",
}
```

```python


# Load dataset
tsw = pd.read_csv("../processed/trimmed_single_words.csv")

# Column containing words
column = "crisis_names"

# Convert to lowercase
tsw[column] = tsw[column].str.lower()

# Replace words using mapping
tsw[column] = tsw[column].replace(replace_map)

# Remove duplicates
new_df = tsw.drop_duplicates(subset=[column])

# Reset index
new_df = new_df.reset_index(drop=True)

# Print result
print(new_df)

# Save new dataset
new_df.to_csv("../processed/normalized_trimmed_single_words.csv", index=False)
```

```python
ntsw = pd.read_csv("../processed/normalized_trimmed_single_words.csv")
ntsw.shape
```

```python
keyword_groups = defaultdict(list)


keywords = ntsw["crisis_names"].dropna().str.lower().unique()

# normalize sorted_df column
sorted_df["crisis_names"] = sorted_df["crisis_names"].str.lower()

# loop through keywords
for word in keywords:

    # regex pattern
    pattern = rf"{re.escape(word)}"

    # grep matching rows
    matches = sorted_df[
        sorted_df["crisis_names"].str.contains(
            pattern,
            case=False,
            regex=True,
            na=False
        )
    ]

    # store as list
    keyword_groups[word] = matches["crisis_names"].tolist()
```

```python
keyword_groups["fire"]
```

creating the keyword_group as datset for weak mapping

```python
import pandas as pd

data = []

for keyword, phrases in keyword_groups.items():

    # remove duplicates
    unique_phrases = list(set(phrases))

    data.append({
        "disaster": keyword,
        "matching_keywords": ", ".join(unique_phrases)
    })

# create dataframe
weak_mapping_df = pd.DataFrame(data)

# preview
print(weak_mapping_df.head())
```
