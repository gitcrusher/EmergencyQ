```python
replace_map ={
    "bomber":"bomb",
    "bombing":"bomb",

    "bushfire":"fire",
    "firefighters":"fire",

    "dead":"death",
    "deadly":"death",

    "floodwaters":"flood",
}
```

```python


# Load dataset
tsw = pd.read_csv("../processed/trimmed_single_words.csv")

# Column containing words
column = "crisis_names"

# Convert to lowercase
tsw[column] = tsw[column].str.lower()

# Replace words using mapping
tsw[column] = tsw[column].replace(replace_map)

# Remove duplicates
new_df = tsw.drop_duplicates(subset=[column])

# Reset index
new_df = new_df.reset_index(drop=True)

# Print result
print(new_df)

# Save new dataset
new_df.to_csv("../processed/normalized_trimmed_single_words.csv", index=False)
```
