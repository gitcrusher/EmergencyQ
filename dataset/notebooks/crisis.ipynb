```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
df = pd.read_csv("../raw/CrisisLexRec.txt", sep="\t", engine="python")
```

```python
sorted_df = df.sort_values(by="crisis_names",ascending = True)
```

```python
sorted_df.head()
```

```python
sorted_df.shape
```

```python
from collections import Counter

all_words = []

for sentence in sorted_df["crisis_names"]:
    words = sentence.lower().split()
    all_words.extend(words)

word_freq = Counter(all_words)

common_words = word_freq.most_common(70)

print(common_words)
```
